# STEP 5: 25 random images from the training data with the label

```python
W_grid= 5
L_grid= 5

fig, axes= plt.subplots(W_grid, L_grid, figsize= (25, 25))
axes= axes.ravel()
n_training= len(X_train)

for i in np.arange(0, W_grid * L_grid):
    index= np.random.randint(0, n_training)
    axes[i].imshow(X_train[index])
    axes[i].set_title(y_train[index], fontsize= 25)
    axes[i].axis("off")
    
plt.subplots_adjust(hspace= 0.5)
```

# STEP 6: Data normalization

```python
X_train= X_train/255
```

```python
X_test= X_test/255
```

```python
X_train
```

```python
X_test
```

# STEP 7: Build the model

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model= Sequential()

cnn_model.add(Conv2D(64, (3, 3), input_shape= (64, 64, 3), activation= "relu"))
cnn_model.add(MaxPooling2D(pool_size= (2, 2)))

cnn_model.add(Dropout(0.2))

cnn_model.add(Conv2D(64, (3, 3), activation= "relu"))
cnn_model.add(MaxPooling2D(pool_size= (2, 2)))

cnn_model.add(Dropout(0.2))

cnn_model.add(Flatten())

cnn_model.add(Dense(activation= "relu", units= 128))# Hidden layer, one of the fc layers
cnn_model.add(Dense(activation= "relu", units= 256))

cnn_model.add(Dense(activation= "sigmoid", units= 1)) # output
```

```python
cnn_model.compile(loss= "binary_crossentropy", optimizer= "Adam", metrics= ["accuracy"] )# when the output is 0 or 1
```

```python
# STEP 8: Train the model
```

```python
history= cnn_model.fit(X_train, y_train, batch_size= 30, epochs= 20, verbose= 1)
```

# STEP 9: Test the model

```python
evaluate= cnn_model.evaluate(X_test, y_test)
```

```python
print("Test_Accuracy: {}".format(evaluate[1]))
```

```python
predicted_class= cnn_model.predict_classes(X_test)
```

```python
predicted_class
```

```python
L_grid= 5
W_grid= 5

fig, axes= plt.subplots(L_grid, W_grid, figsize= (25, 25))
axes= axes.ravel()

for i in np.arange(0, L_grid * W_grid):
    axes[i].imshow(X_test[i])
    axes[i].set_title("Guess= {}\n True= {}".format(predicted_class[i], y_test[i]), fontsize= 25)
    axes[i].axis("off")
plt.subplots_adjust(hspace= 0.5)
plt.subplots_adjust(wspace= 0.5)
```

# STEP 10: Result shown in Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm= confusion_matrix(y_test, predicted_class)
plt.figure(figsize= (10, 10))
sns.heatmap(cm, annot= True)
```

```python
print(classification_report(y_test, predicted_class))
```
